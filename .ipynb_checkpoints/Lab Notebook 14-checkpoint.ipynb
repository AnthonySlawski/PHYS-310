# Lab 14: Flavours of Boosting and Feature Importance

In this notebook, we build a photometric redshift estimator using various boosting methods: AdaBoost and various flavors of Gradient Boosted Trees (GBM, HistGBM, and XGBoost). We also look at using RandomizedSearchCV in order to improve our exploration of parameter space.

Our goal is to estimate photometric redshifts starting from observations of galaxy magnitudes in six different photometric bands (u, g, r, i, z, y).

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib
import matplotlib.pyplot as plt
from joblib import parallel_backend

pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_colwidth', 100)

font = {'size'   : 16}
matplotlib.rc('font', **font)
matplotlib.rc('xtick', labelsize=14) 
matplotlib.rc('ytick', labelsize=14) 
#matplotlib.rcParams.update({'figure.autolayout': True})
matplotlib.rcParams['figure.dpi'] = 100
```

```python
from sklearn import metrics
from sklearn.model_selection import cross_validate, cross_val_score, KFold, cross_val_predict, GridSearchCV,RandomizedSearchCV
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier, AdaBoostRegressor, GradientBoostingRegressor,RandomForestRegressor

# read in pre-processed data from Lab 13
sel_features = pd.read_csv('/Users/erinbennett/Downloads/sel_features.csv', sep = '\t')
sel_target = pd.read_csv('/Users/erinbennett/Downloads/sel_target.csv')
```

## Step 1: AdaBoost again 

In Lab 13, we showed that for AdaBoostRegressor, stacking learners that are too weak doesn't help. Let's do a more thorough optimization of hyperparameters for AdaBoost. 

First, look at the model hyperparameters using model.get_params. As estimator use DecisionTreeRegressor(). Set up a **GridSearchCV** and find the best model within the suggested parameters.  

'estimator__max_depth':[6,10,None], 'loss':['linear','square'], 'n_estimators':[20,50,100], 'learning_rate': [0.3,0.5,1.0]

You can use 3-fold CV. Use n_jobs=-1 to request all resources. If the run time turns out to be too long on your machine, use **RandomizedSearchCV** with n_iter=50 or 100 instead. Normally we would only use the randomized search in higher dimensional parameter space, but it should be sufficient here. You can increase n_inter according to your patience threshold!

What are the test scores, optimal parameters and outlier fraction of the winning model? 

```python
param_grid = {
    'base_estimator__max_depth': [6, 10, None],
    'loss': ['linear', 'square'],
    'n_estimators': [20, 50, 100],
    'learning_rate': [0.3, 0.5, 1.0]
}

ada_boost = AdaBoostRegressor(base_estimator=DecisionTreeRegressor())

grid_search = GridSearchCV(ada_boost, param_grid, cv=3, n_jobs=-1)
grid_search.fit(sel_features, sel_target)
```

```python
best_model = grid_search.best_estimator_

best_params = grid_search.best_params_
test_scores = grid_search.cv_results_['mean_test_score']

best_model.fit(sel_features, sel_target)
pred = best_model.predict(sel_features)
outlier_fraction = np.sum(np.abs((sel_target.values.flatten() - pred) / (1 + sel_target.values.flatten())) > 0.15) / len(pred)


print("Test scores:", test_scores)
print("\nMean Test scores", np.mean(test_scores))
print("\nOptimal parameters:", best_params)
print("\nOutlier Fraction:", outlier_fraction)
```

## Step 2: Gradient Boosted Trees

Let's repeat the analysis above for three GB variants:
1. GradientBoostingRegressor

Suggested parameters: 'max_depth':[6,10,None], 'loss':['squared_error','absolute_error'], 'n_estimators':[20,50,100], 'learning_rate': [0.1,0.3,0.5]

2. HistGradientBoostingRegressor

HistGradientBoostingRegressor (inspired by LightGBM) works by binning the features into integer-valued bins (the default value is 256, but this parameter can be adjusted; note however that 256 is the maximum!), which greatly reduces the number of splitting points to consider, and results in a vast reduction of computation time, especially for large data sets.

Suggested paramters: 'max_depth':[6,10,None], 'loss':['squared_error','absolute_error'], 'max_iter':[20,50,100], 'learning_rate': [0.1,0.3,0.5]

3. XGBRegressor. The latter requires installation of the python package xgboost. 

XGBoost stands for “Extreme Gradient Boosting”. It is sometimes known as "regularized" GBM, as it has a default regularization term on the weights of the ensemble, and is more robust to overfitting. It has more flexibility in defining weak learners, as well as the objective (loss) function (note that this doesn't apply to the base estimators, e.g. how splits in trees are chosen, but on the loss that is used to compute pseudoresiduals and gradients).

Suggested parameters: 'max_depth':[6,10,None], 'n_estimators':[50,100,200], 'learning_rate': [0.1, 0.3,0.5], 'objective':['reg:squarederror','reg:squaredlogerror']

What are the test scores, optimal parameters and outlier fraction for the winning models in each case? 

```python
#====GBR======
param_grid_gb = {
    'max_depth': [6, 10, None],
    'loss': ['squared_error', 'absolute_error'],
    'n_estimators': [20, 50, 100],
    'learning_rate': [0.1, 0.3, 0.5]
}

gb = GradientBoostingRegressor()
grid_search_gb = GridSearchCV(gb, param_grid_gb, cv=3, n_jobs=-1)
grid_search_gb.fit(sel_features, sel_target)


best_model_gb = grid_search_gb.best_estimator_

best_params_gb = grid_search_gb.best_params_
test_scores_gb = grid_search_gb.cv_results_['mean_test_score']

best_model_gb.fit(sel_features, sel_target)
pred_gb = best_model_gb.predict(sel_features)
outlier_fraction_gb = np.sum(np.abs((sel_target.values.flatten() - pred_gb) / (1 + sel_target.values.flatten())) > 0.15) / len(pred_gb)


print("Test scores:", test_scores_gb)
print("\nMean Test scores", np.mean(test_scores_gb))
print("\nOptimal parameters:", best_params_gb)
print("\nOutlier Fraction:", outlier_fraction_gb)
```

```python
#====HIST GBR======
from sklearn.ensemble import HistGradientBoostingRegressor

param_grid_hist = {
    'max_depth': [6, 10, None],
    'loss': ['squared_error', 'absolute_error'],
    'max_iter': [20, 50, 100],
    'learning_rate': [0.1, 0.3, 0.5]
}

hist = HistGradientBoostingRegressor()
grid_search_hist = GridSearchCV(hist, param_grid_hist, cv=3, n_jobs=-1)
grid_search_hist.fit(sel_features, sel_target)

best_model_hist = grid_search_hist.best_estimator_
best_params_hist = grid_search_hist.best_params_
test_scores_hist = grid_search_hist.cv_results_['mean_test_score']

best_model_hist.fit(sel_features, sel_target)
pred_hist = best_model_hist.predict(sel_features)
outlier_fraction_hist = np.sum(np.abs((sel_target.values.flatten() - pred_hist) / (1 + sel_target.values.flatten())) > 0.15) / len(pred_hist)

print("Test scores:", test_scores_hist)
print("\nMean Test scores", np.mean(test_scores_hist))
print("\nOptimal parameters:", best_params_hist)
print("\nOutlier Fraction:", outlier_fraction_hist)
```

```python
#====XGBR======
from xgboost import XGBRegressor

param_grid_xgb = {
    'max_depth': [6, 10, None],
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.1, 0.3, 0.5],
    'objective': ['reg:squarederror', 'reg:squaredlogerror']
}

xgb = XGBRegressor()
grid_search_xgb = GridSearchCV(xgb, param_grid_xgb, cv=3, n_jobs=-1)
grid_search_xgb.fit(sel_features, sel_target)

best_model_xgb = grid_search_xgb.best_estimator_
best_params_xgb = grid_search_xgb.best_params_
test_scores_xgb = grid_search_xgb.cv_results_['mean_test_score']

best_model_xgb.fit(sel_features, sel_target)
pred_xgb = best_model_xgb.predict(sel_features)
outlier_fraction_xgb = np.sum(np.abs((sel_target.values.flatten() - pred_xgb) / (1 + sel_target.values.flatten())) > 0.15) / len(pred_xgb)

print("Test scores:", test_scores_xgb)
print("\nMean Test scores", np.mean(test_scores_xgb))
print("\nOptimal parameters:", best_params_xgb)
print("\nOutlier Fraction:", outlier_fraction_xgb)
```

## Step 3: Feature importance

Although one of the main goals of ML is to be able to make predictions as accurate as possible, it is equally desirable to gain insight into why the machine makes certain decisions. To this end, we would like to be able to identify those features that are most important in the classification or regression process. This is sometimes called "interpretable machine learning".

Our new ensemble methods all have a property .feature_importances_. What does this property give you in the case of a random forest?

For the RandomForestRegressor, the AdaBoostRegressor, and the XGBRegressor, find those importances, rank them in order of importance and make a bar chart for each regressor that plots the importances of the six features in our redshift dataset, starting with the most important one. To this end, just fit the models to the full dataset, no split test/train needed. You can use the optimized parameters you found above. Do you get the same answer from all three regressors?

#### The features change importance slightly with all three regressors, but keep the same hierarchy

```python
from sklearn.ensemble import RandomForestRegressor

best_params_rf = {'max_depth': None, 'n_estimators': 100}

rf_regressor = RandomForestRegressor(**best_params_rf)
rf_regressor.fit(sel_features, sel_target.values.ravel())

feature_importances_rf = rf_regressor.feature_importances_


sorted_indices_rf = np.argsort(feature_importances_rf)[::-1]
sorted_importances_rf = feature_importances_rf[sorted_indices_rf]
sorted_features_rf = sel_features.columns[sorted_indices_rf]


plt.figure(figsize=(10, 6))
plt.bar(sorted_features_rf, sorted_importances_rf)
plt.title('Feature Importances - RandomForestRegressor')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.xticks(rotation=45)
plt.show()
```

```python
best_params_ada = {'base_estimator': DecisionTreeRegressor(max_depth=None), 'learning_rate': 1.0, 'loss': 'linear', 'n_estimators': 100}

ada_regressor = AdaBoostRegressor(**best_params_ada)
ada_regressor.fit(sel_features, sel_target.values.ravel())

feature_importances_ada = ada_regressor.feature_importances_


sorted_indices_ada = np.argsort(feature_importances_ada)[::-1]
sorted_importances_ada = feature_importances_ada[sorted_indices_ada]
sorted_features_ada = sel_features.columns[sorted_indices_ada]


plt.figure(figsize=(10, 6))
plt.bar(sorted_features_ada, sorted_importances_ada)
plt.title('Feature Importances - AdaBoostRegressor')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.xticks(rotation=45)
plt.show()
```

```python


best_params_xgb = {'max_depth': None, 'n_estimators': 100, 'learning_rate': 0.3, 'objective': 'reg:squarederror'}

xgb_regressor = XGBRegressor(**best_params_xgb)
xgb_regressor.fit(sel_features, sel_target.values.ravel())

feature_importances_xgb = xgb_regressor.feature_importances_


sorted_indices_xgb = np.argsort(feature_importances_xgb)[::-1]
sorted_importances_xgb = feature_importances_xgb[sorted_indices_xgb]
sorted_features_xgb = sel_features.columns[sorted_indices_xgb]


plt.figure(figsize=(10, 6))
plt.bar(sorted_features_xgb, sorted_importances_xgb)
plt.title('Feature Importances - XGBRegressor')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.xticks(rotation=45)
plt.show()
```

## Step 4: Correlations among features

It is very common for the features to be correlated amonst each other. This can complicate the interpretation of feature importance, since two highly correlated features might receive equal weight. Correlated features also introduce reduncancy in the overall model so it is desirable to develop strategies to remove them.

1. Determine the cross-correlations between the six features in our dataset. There are several options to do this, in pandas we could just use the .corr() method. Find the pair of features that is most correlated.

```python
sel_features.corr()
```

##### The most correlated pair is z_apercor and i_apercor

2. From these two features, which one is less correlated with the target variable? Eliminate that feature and redo the feature ranking for the three regressors with this new model that now only has 5 features. Also compute a cross-validated score for these smaller models, are they worse than for the full model? What do you think? 

#### The z_apercor is less correlated to the target
#### The cv scores are slightly better but about the same, eliminating a the feature slightly improved the results, but not by much.

```python
test_z = pd.concat([sel_features['z_apercor'], sel_target['zhelio']], axis=1)
test_i = pd.concat([sel_features['i_apercor'], sel_target['zhelio']], axis=1)

print(test_z.corr())
print(test_i.corr())

# i_apercor is more correlated to the target

sel_features_drop = sel_features.drop(columns=['z_apercor'])
```

```python
# Random Tree Regressor

rf_regressor = RandomForestRegressor()
rf_regressor.fit(sel_features_drop, sel_target.values.ravel())

feature_importances_rf = rf_regressor.feature_importances_


sorted_indices_rf = np.argsort(feature_importances_rf)[::-1]
sorted_importances_rf = feature_importances_rf[sorted_indices_rf]
sorted_features_rf = sel_features_drop.columns[sorted_indices_rf]

scores_rf = cross_validate(rf_regressor, sel_features_drop, sel_target)
mean_score_rf = np.mean(scores_rf['test_score'])
print("Mean CV Test Score:", mean_score_rf)

plt.figure(figsize=(10, 6))
plt.bar(sorted_features_rf, sorted_importances_rf)
plt.title('Feature Importances - RandomForestRegressor')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.xticks(rotation=45)
plt.show()
```

```python
#AdaBoost
best_params_ada = {'base_estimator': DecisionTreeRegressor(max_depth=None), 'learning_rate': 1.0, 'loss': 'linear', 'n_estimators': 100}

ada_regressor = AdaBoostRegressor(**best_params_ada)
ada_regressor.fit(sel_features_drop, sel_target.values.ravel())

feature_importances_ada = ada_regressor.feature_importances_


sorted_indices_ada = np.argsort(feature_importances_ada)[::-1]
sorted_importances_ada = feature_importances_ada[sorted_indices_ada]
sorted_features_ada = sel_features_drop.columns[sorted_indices_ada]

scores_ada = cross_validate(ada_regressor, sel_features_drop, sel_target)
mean_score_ada = np.mean(scores_ada['test_score'])
print("Mean CV Test Score:", mean_score_ada)

plt.figure(figsize=(10, 6))
plt.bar(sorted_features_ada, sorted_importances_ada)
plt.title('Feature Importances - AdaBoostRegressor')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.xticks(rotation=45)
plt.show()
```

```python
# XBGRegressor
best_params_xgb = {'max_depth': None, 'n_estimators': 100, 'learning_rate': 0.3, 'objective': 'reg:squarederror'}

xgb_regressor = XGBRegressor(**best_params_xgb)
xgb_regressor.fit(sel_features_drop, sel_target.values.ravel())

feature_importances_xgb = xgb_regressor.feature_importances_


sorted_indices_xgb = np.argsort(feature_importances_xgb)[::-1]
sorted_importances_xgb = feature_importances_xgb[sorted_indices_xgb]
sorted_features_xgb = sel_features_drop.columns[sorted_indices_xgb]

scores_xgb = cross_validate(xgb_regressor, sel_features_drop, sel_target)
mean_score_xgb = np.mean(scores_xgb['test_score'])
print("Mean CV Test Score:", mean_score_xgb)


plt.figure(figsize=(10, 6))
plt.bar(sorted_features_xgb, sorted_importances_xgb)
plt.title('Feature Importances - XGBRegressor')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.xticks(rotation=45)
plt.show()
```
