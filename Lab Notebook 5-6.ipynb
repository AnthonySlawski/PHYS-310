# Part 5: Cross-validation
Implement cross-validation on both models for the three metrics accuracy, precision, and recall with stratified k-folds, random shuffling, and 10 splits, random_state=10. What is k-fold validation, and what happens when stratification is applied? Report the mean CV score and the standard deviation. 

```python
from sklearn.model_selection import cross_validate

cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=10)

precision_scorer = make_scorer(precision_score)
recall_scorer = make_scorer(recall_score)

knn_cv_accuracy = cross_val_score(model_knn, final_features, targets, cv=cv, scoring='accuracy')
knn_cv_precision = cross_val_score(model_knn, final_features, targets, cv=cv, scoring=precision_scorer)
knn_cv_recall = cross_val_score(model_knn, final_features, targets, cv=cv, scoring=recall_scorer)
knn_scores = cross_val_score(model_knn, Xtest, ytest, cv=cv)
print(knn_scores.mean())
print(knn_scores.std())

dt_cv_accuracy = cross_val_score(model_dt, final_features, targets, cv=cv, scoring='accuracy')
dt_cv_precision = cross_val_score(model_dt, final_features, targets, cv=cv, scoring=precision_scorer)
dt_cv_recall = cross_val_score(model_dt, final_features, targets, cv=cv, scoring=recall_scorer)

dt_scores = cross_val_score(model_dt, Xtest, ytest, cv=cv)
print(dt_scores.mean())
print(dt_scores.std())
```

#### k fold validation is dividing a set of observations into k groups of equal size, then training and evaluating the model k times, and averaging the performance. 

#### Stratification makes sure that each group has a similar mix of different types of examples when we split our data into groups for testing and training. This helps to get more fair evaluations

# Part 6: Confusion Matrix

Construct the confusion matrices for the two classifiers from above. Please write code from scratch to compute the matrix elements, but feel free to check it against the function provided by sklearn.metrics. Use **cross_val_predict** to assemble the predictions of several folds. You can then use **ConfusionMatrixDisplay()** to visualize, or write your own plotting routine.

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

def my_confusion_matrix(y_true, y_pred):
    true_positive = np.sum((y_true == 1) & (y_pred == 1))
    true_negative = np.sum((y_true == 0) & (y_pred == 0))
    false_positive = np.sum((y_true == 0) & (y_pred == 1))
    false_negative = np.sum((y_true == 1) & (y_pred == 0))
    return true_positive, true_negative, false_positive, false_negative

# KNN
y_pred_knn = cross_val_predict(model_knn, Xtest, ytest, cv=cv)

# My own confusion matrix
tp_knn, tn_knn, fp_knn, fn_knn = my_confusion_matrix(ytest, y_pred_knn)

# Sklearn confusion matrix
sk_matrix_knn = confusion_matrix(ytest, y_pred_knn)

disp = ConfusionMatrixDisplay(sk_matrix_knn)
disp.plot(cmap=plt.cm.Blues)

# Decision Tree
y_pred_dt = cross_val_predict(model_dt, Xtest, ytest, cv=cv)

# My own confusion matrix
tp_dt, tn_dt, fp_dt, fn_dt = my_confusion_matrix(ytest, y_pred_dt)

# Sklearn confusion matrix
sk_matrix_dt = confusion_matrix(ytest, y_pred_dt)

disp = ConfusionMatrixDisplay(sk_matrix_dt)
disp.plot(cmap=plt.cm.Greens)

plt.show()
```

# Part 8: Learning curves

With the help of the module **learning_curve** from sklearn.model_selection, construct the learning curves for both classifiers as shown in Fig. 3.7.

```python
from sklearn.model_selection import learning_curve

train_sizes=[0.2,0.3, 0.6,0.7, 0.9]
train_size_abs, train_scores, test_scores = learning_curve(model_knn, final_features, targets,cv=cv, train_sizes=[0.2,0.3, 0.6,0.7, 0.9])

avg_train_scores = []
std_train_scores = []
avg_test_scores = []
std_test_scores = []
for i in train_scores:
    avg_train_scores.append(np.mean(i))
    
for i in train_scores:
    std_train_scores.append(np.std(i))

for i in test_scores:
    avg_test_scores.append(np.mean(i))
    
for i in test_scores:
    std_test_scores.append(np.std(i))
```
