# Linear Regression to learn the coupling constants of the one-dimensional Ising model

Let us consider a one dimensional chain of $L$ (classical) spins $S_j=\pm 1$. Their interaction with other spins $S_k$ is described by a coupling matrix $J_{j,k}$ so that in general, every spin can interact with all other spins. The energy for a spin confiuration $\boldsymbol{S}^i$ (aka the "Hamiltonian") can be written as  

$$
H_\mathrm{Ising}[\boldsymbol{S}^i] = - \sum_{j=1}^L \sum_{k=1}^L J_{j,k}S_{j}^iS_{k}^i.
$$

This is the famous one-dimensional Ising model from Statistical Mechanics.

### Learning the Ising model ###

Suppose your machine learning friend drew a large number of spin configurations, and computed their Ising energies. Then, without telling you about the above Hamiltonian, they handed you a data set of $i=1\ldots n$ points of the form $\{(H[\boldsymbol{S}^i],\boldsymbol{S}^i)\}$. Your task is to learn the non-local coupling strengths $J_{j,k}$ using Linear regression techniques. Importantly, this model is linear in ${\mathbf J}$ which makes it possible to use linear regression.

To apply linear regression, we would like to recast this model in the form
$$
H_\mathrm{model}^i \equiv \mathbf{X}^i \cdot \mathbf{J},
$$

where the vectors $\mathbf{X}^i$ represent all two-body interactions $\{S_{j}^iS_{k}^i \}_{j,k=1}^L$, and the index $i$ runs over the samples in the data set. To make the analogy complete, we can also represent the dot product by a single index $p = \{j,k\}$, i.e. $\mathbf{X}^i \cdot \mathbf{J}=X^i_pJ_p$. Note that the regression model does not include the minus sign, so we expect to learn negative $J$'s.

There are two of the above mentioned datasets, namely *states_system1.txt*  - *energies_system1.txt* and another pair of such files for a second system. Both datasets contain 10,000 realizations of a 1d spin chain of $L=30$. Your friend was very kind and already computed for you all $L\times L=900$ possible spin-spin interactions $S_jS_k$ and listed them for each of the 10,000 realizations.

1. Load the first dataset. Set up linear regression, choose a suitable evalutation metric, compute the learning curve in the range of 10-2000 samples and interpret the result. What is the minimium number of samples needed, and does adding more samples improve the model?

```python
states=np.loadtxt('states_system1.txt')
energies=np.loadtxt('energies_system1.txt')
```

```python
X_train,X_test,y_train,y_test=train_test_split(states,energies,test_size=0.2,random_state=42)

model=LinearRegression()

sizes=np.arange(10,2001,10)
mse_scores=[]

for size in sizes:
    model.fit(X_train[:size],y_train[:size])
    y_pred=model.predict(X_test)
    mse=mean_squared_error(y_test, y_pred)
    mse_scores.append(mse)

plt.plot(sizes,mse_scores,marker='o')
plt.xlabel('Number of Samples')
plt.ylabel('Mean Squared Error')
plt.title('Learning Curve for Linear Regression')
```

#### The minimum amount of samples is around 450. The model is not improved by adding more samples beyond the minimum number of samples.

2. Create a reasonable split between training and testing datasets, fit the model and visualize the matrix $J_{j,k}$ of coupling constants. You need to reshape the model coefficients into a $L\times L$ array. Can you see a pattern?

```python
model.fit(X_train[:500],y_train[:500])

coupling_matrix=model.coef_.reshape((30,30))

plt.imshow(coupling_matrix)
plt.colorbar(label='Coupling Constants')
plt.title('Coupling Constants Matrix')
plt.xlabel('j')
plt.ylabel('k')
```

#### The constants are ~0 except for at (0,30), (30,0), and just either side of the line j=k.

3. Compute the distributions of coupling constants (or histograms) for several sample sizes, starting at 100 samples, and plot them together. What do you see?

```python
sizes=[100,200,300,400,500]

fig,ax=plt.subplots(len(sizes),1,figsize=(8,2*len(sizes)))

for i, size in enumerate(sizes):
    X_train,X_test,y_train,y_test=train_test_split(states,energies,test_size=0.2,random_state=42)
    X_train_s=X_train[:size]
    y_train_s=y_train[:size]

    model=LinearRegression()
    model.fit(X_train_s,y_train_s)

    coupling_matrix=model.coef_.reshape((30,30))

    ax[i].hist(coupling_matrix.flatten(),bins=30)
    ax[i].set_title(('Sample Size: %d'%size))
    ax[i].set_xlabel('Coupling Constants')
    ax[i].set_ylabel('Frequency')
plt.tight_layout()
```

#### As the sample size increases, the distribution converges towards 0, however the outlier at -1 remains.

5. If you find your linear regression analysis above not entirely conclusive, you could consider trying a regularized model, such as Lasso. Try this model with the regularization parameter $\alpha$ in the range of 0.01-0.00001. Do any answers beome clearer? 

```python
states=np.loadtxt('states_system1.txt')
energies=np.loadtxt('energies_system1.txt')

alphas=[0.01,0.001,0.0001,0.00001]

fig,ax=plt.subplots(len(alphas),1,figsize=(8,2*len(alphas)))

for i, alpha in enumerate(alphas):
    X_train,X_test,y_train,y_test=train_test_split(states,energies,test_size=0.2,random_state=42)
    X_train_s=X_train[:size]
    y_train_s=y_train[:size]

    model=Lasso(alpha=alpha,max_iter=10000)
    model.fit(X_train_s,y_train_s)

    coupling_matrix=model.coef_.reshape((30,30))

    ax[i].hist(coupling_matrix.flatten(),bins=30)
    ax[i].set_title(('Regularization Parameter: %.5f'%alpha))
    ax[i].set_xlabel('Coupling Constants')
    ax[i].set_ylabel('Frequency')
plt.tight_layout()
```
